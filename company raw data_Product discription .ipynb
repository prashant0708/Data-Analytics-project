# for Merging renaming the pid to PID 

```python
df2.rename(columns={'pId':"PID"},inplace=True)
df2
```

# merging df1(Product description) and df2(Product proformance Table)

```python
df = pd.merge(df1, df2, on="PID")
```

```python
##deleting the p name from the table
del df['pName']
```

```python
df
```

```python
df.describe()
```

# Ploting bar grapth to observe which fabric user has ordered more .

```python
fig=plt.figure(figsize=(100,50))
plt.rcParams.update({'font.size': 40})
plt.bar(df['Fabric'],df['orders'])
plt.xlabel("fabric")
plt.ylabel("order")
plt.title("fabric vs order")
plt.show()
```

# Observation 
user has ordered Cotton Spandex more, if we add more Cotton Spandex product may order frequency increase 

```python
fig=plt.figure(figsize=(85,40))
plt.rcParams.update({'font.size':40})
plt.bar(df['Fabric'],df['clicks'])
plt.xlabel('Fabric')
plt.ylabel('clicks')
plt.title('Fabric vs clicks')
```

# Observation 
# user has click more on cotton fabric but order Cotton Spandex more 


```python
### bar grapgh of catagory and order to see which category has been order more 
fig=plt.figure(figsize=(20,8))
plt.rcParams.update({"font.size":10})
plt.bar(df['Category'],df['orders'],width=0.4)
plt.xlabel('Catagory')
plt.ylabel('order')
plt.title('Cat vs order')
```

# Observation
